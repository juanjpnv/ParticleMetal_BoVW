## Base de dados

O dataset contém imagens sintéticas que simulam a visualização de materiais particulados em um Microscópio Eletrônico de Varredura (sigla em inglês, SEM). As imagens foram renderizadas no software Blender.

As classes se diferenciam entre si pela Distribuição de Tamanho das Partículas (do inglês, PSD). Cada partícula é considerada esférica, e o tamanho de cada uma varia randomicamente ao redor do valor de PSD da classe.

* 2048 imagens sintéticas, que simulam imagens de microscópio eletrônico de varredura.
* 8 classes (256 imagens cada), com diferentes PSD
* Descrição das caracteristicas tridimensionais dos 8 tipos de estruturas
* Descrição das estruturas que estão sendo representadas

````python
def dataframe_from_folder(raiz):
  """
    Retorna um dataframe com as imagens.

    Parâmetros
    ----------
    raiz : str 
        Path da pasta raiz

    Return
    ------
    df : dataframe
        DataFrame a classe e o
        path da imagem.
  """
  data = []
  for classe in tqdm(os.listdir(raiz)):
    for img in os.listdir(f'{raiz}/{classe}'):
      img_path = f'{raiz}/{classe}/{img}'

      if '.gif' not in img_path: # Evita adicionar gifs na lista de imagens
        linha = [classe, img_path]
        data.append(linha)

  df = pd.DataFrame(data, columns=['classe', 'path'])

  return df
````

## Classes

````python
class BagOfVisualWords:

  def __init__(self, n_clusters, image_descriptor=cv2.xfeatures2d.SIFT_create(), mask=None, cluster_verbose=3, cluster_n_jobs=-1):
    # Variáveis recebidas do usuário
    self.mask = mask
    self.n_clusters = n_clusters

    # Instanciando os objetos úteis
    self.image_descriptor = image_descriptor # descritor de imagem. Pode ser SIFT, ORB, SURF, BRISK ou outro semelhante.
    self.cluster = KMeans(n_clusters, verbose=cluster_verbose, n_jobs=cluster_n_jobs)

    # Criando variáveis internas
    self.codebook = []

  # Funções
  def feature_extraction(self, IMGS):
    """
      Utiliza o descritor de imagem para extrair os pontos chaves das imagens, 
      e prepara os dados para serem inseridos no processo de clusterização.
      
      Parametros
      ----------
      IMGS: Conjunto de imagens.

      return
      ------
      count: contagem.
    """

    print('Extraindo características')
    # extrai as características da primeira imagem
    img = IMGS[0]
    kps, des = self.image_descriptor.detectAndCompute(img, self.mask)

    # Com o resultado da primeira imagem, é iniciada a pilha de características
    descriptors = des

    # E empilha as caracteristivas das demais imagens nesta pilha.
    for img in tqdm(IMGS[1:]):
      kps, des = self.image_descriptor.detectAndCompute(img, self.mask)
      descriptors = np.vstack((descriptors, des))
    
    return descriptors.astype(np.float64)

  def fit(self, IMGS, y=None):
    """
      Efetua o treinamento do BoVW. Ao final deste processo deve ter sido gerado o codebook.

      IMGS: Conjunto de imagens.
      y: [None]. Só está aqui pra facilitar compatibilidade com o sklearn. 
    """
    descriptors = self.feature_extraction(IMGS)
    print("Efetuando o cluster. Isso pode levar um tempo, e não têm barra de progresso...", "")
    self.cluster.fit(descriptors)
    self.codebook = self.cluster.cluster_centers_

  def _get_bovw_feat(self, img):
    """
      Conta quantas vezes cada palavra do codebook aparece nos pontos chaves extraídos pelo método
      de extração primário.

      Parametros
      ----------
      img: Uma imagem a ser trabalhada

      return
      ------
      count: contagem.
    """
    kps, des = self.image_descriptor.detectAndCompute(img, self.mask)
    x = self.cluster.predict(des)

    count = np.zeros(self.n_clusters)
    for i in x:
      count[i] += 1

    return count

  def transform(self, IMGS, y=None):
    """
      Recebe o conjunto de imagens e aplica BoVW, retornando as features extraídas.

      Parametros:
      IMGS: conjunto de imagens
      
      Retorna
      -------
      X: Image features
      y: Labels
      
    """
    X = [self._get_bovw_feat(img) for img in tqdm(IMGS)]
    X = np.asarray(X)

    return X, y
````

# BOVW com imagens coloridas

## Pegando os dados

````python
data_images = dataframe_from_folder(pasta_raiz)
data_images.head()
````

````python
IMGS = [cv2.imread(path) for path in tqdm(data_images['path'])]
y = data_images['classe']
````

## Separação dos dados
Separando treino, teste e validação

````python
IMGS_train_full, IMGS_test, y_train_full, y_test = train_test_split(IMGS, y,
                                                                    test_size=0.5, 
                                                                    random_state=42,
                                                                    shuffle=True,
                                                                    stratify=y)

IMGS_train, IMGS_valid, y_train, y_valid = train_test_split(IMGS_train_full, 
                                                            y_train_full,
                                                            train_size=0.625, 
                                                            random_state=42,
                                                            shuffle=True,
                                                            stratify=y_train_full)

print(f"Train: {len(IMGS_train)}")
print(f"Valid: {len(y_valid)}")
print(f"Test: {len(y_test)}")
````

## Aplicando BOVW Coloridas.

Apenas o código básico de Bag of Words, sem nenhum preprocessamento ou tratamento sobre a imagem

````python
bovw = BagOfVisualWords(n_clusters=100)
````

````python
inicio_fit = time()

bovw.fit(IMGS_train)

tempo_fit_BOVW_simples_colorido = time()-inicio_fit
````

````python
inicio_transform = time()

X_train, y_train = bovw.transform(IMGS_train, y_train)
X_valid, y_valid = bovw.transform(IMGS_valid, y_valid)
X_test, y_test = bovw.transform(IMGS_test, y_test)

tempo_transform_BOVW_simples_colorido = time()-inicio_transform

print(f'Tempo de treino BOVW: {tempo_fit_BOVW_simples_colorido}')
print(f'Tempo de transformação de dados BOVW: {tempo_transform_BOVW_simples_colorido}')
````

### Aprendizagem de máquina

#### SVM

````python
# Os hiperparâmetros de busca serão os mesmos para os 4 classificadores, 
# independente do modelo de extração de característica.
param_grid_SVM = {'C': [1, 10, 100, 1000],
                  'kernel': ['linear', 'rbf', 'sigmoid'],
                  'gamma': [0.001, 0.0001, 'auto', 'scale'],
                  'class_weight':['balanced', None],
                  'decision_function_shape': ['ovo']}

grid_svm = GridSearchCV(SVC(), 
                        param_grid_SVM, 
                        scoring = ['accuracy', 'f1_macro'],
                        cv = 5,
                        refit = 'f1_macro',
                        verbose=True)

grid_svm.fit(X_train, y_train)

y_valid_pred = grid_svm.predict(X_valid)

plt.title('Validação SVM')
cf_matrix = confusion_matrix(y_valid, y_valid_pred)
sns.heatmap(cf_matrix, annot=True, fmt='g', cmap='Blues')

print(classification_report(y_valid, y_valid_pred))
````

````python
y_test_pred = grid_svm.predict(X_test)

plt.title('Teste SVM')
cf_matrix = confusion_matrix(y_test, y_test_pred)
sns.heatmap(cf_matrix, annot=True, fmt='g', cmap='Blues')

print(classification_report(y_test, y_test_pred))
````

#### kNN

````python
# Os hiperparâmetros de busca serão os mesmos para os 4 classificadores, 
# independente do modelo de extração de característica.
from sklearn.neighbors import KNeighborsClassifier

param_grid_KNN = {'n_neighbors':(3,5,7),
                  'weights':('uniform','distance'),
                  'algorithm':('ball_tree','kd_tree','brute'),
                  'leaf_size':(10,20,30,40,50)}

grid_knn = GridSearchCV(KNeighborsClassifier(), 
                        param_grid_KNN, 
                        scoring = ['accuracy', 'f1_macro'],
                        cv = 5,
                        refit = 'f1_macro',
                        verbose=True)

grid_knn.fit(X_train, y_train)

y_valid_pred = grid_knn.predict(X_valid)
plt.title('Validação kNN')
cf_matrix = confusion_matrix(y_valid, y_valid_pred)
sns.heatmap(cf_matrix, annot=True, fmt='g', cmap='Blues')

print(classification_report(y_valid, y_valid_pred))
````

````python
y_test_pred = grid_knn.predict(X_test)

plt.title('Teste kNN')
cf_matrix = confusion_matrix(y_test, y_test_pred)
sns.heatmap(cf_matrix, annot=True, fmt='g', cmap='Blues')

print(classification_report(y_test, y_test_pred))
````

#### RF

````python
# Os hiperparâmetros de busca serão os mesmos para os 4 classificadores, 
# independente do modelo de extração de característica.
from sklearn.ensemble import RandomForestClassifier

param_grid_RF = {'n_estimators': (100, 300, 500),
                 'criterion': ('gini', 'entropy'),
                 'max_features': ('auto', 'sqrt', 'log2'),
                 'ccp_alpha': (0.0, .5, 1.)}

grid_forest = GridSearchCV(RandomForestClassifier(), 
                        param_grid_RF, 
                        scoring = ['accuracy', 'f1_macro'],
                        cv = 5,
                        refit = 'f1_macro',
                        verbose=True)

grid_forest.fit(X_train, y_train)

y_valid_pred = grid_forest.predict(X_valid)
plt.title('RF validação')
cf_matrix = confusion_matrix(y_valid, y_valid_pred)
sns.heatmap(cf_matrix, annot=True, fmt='g', cmap='Blues')

print(classification_report(y_valid, y_valid_pred))
````

````python
y_test_pred = grid_forest.predict(X_test)

plt.title('Teste RF')
cf_matrix = confusion_matrix(y_test, y_test_pred)
sns.heatmap(cf_matrix, annot=True, fmt='g', cmap='Blues')

print(classification_report(y_test, y_test_pred))
````

````python
# Os hiperparâmetros de busca serão os mesmos para os 4 classificadores, 
# independente do modelo de extração de característica.
param_grid_SVM = {'C': [1, 10, 100, 1000],
                  'kernel': ['linear', 'rbf', 'sigmoid'],
                  'gamma': [0.001, 0.0001, 'auto', 'scale'],
                  'class_weight':['balanced', None],
                  'decision_function_shape': ['ovo']}

grid_svm_gray = GridSearchCV(SVC(), 
                        param_grid_SVM, 
                        scoring = ['accuracy', 'f1_macro'],
                        cv = 5,
                        refit = 'f1_macro',
                        verbose=True)

grid_svm_gray.fit(X_train, y_train)

y_valid_pred = grid_svm_gray.predict(X_valid)

print('\nVALIDAÇÃO SVM - Simples, imgs ciza')
print(classification_report(y_valid, y_valid_pred))
plt.figure()
plt.title('Validação SVM')
cf_matrix = confusion_matrix(y_valid, y_valid_pred)
sns.heatmap(cf_matrix, annot=True, fmt='g', cmap='Blues' 
            xticklabels=grid_svm_gray.classes_,
            yticklabels=grid_svm_gray.classes_)


print('\TESTE SVM - Simples, imgs ciza')
y_test_pred = grid_svm_gray.predict(X_test)
cf_matrix = confusion_matrix(y_test, y_test_pred)
plt.figure()
plt.title('Teste SVM')
sns.heatmap(cf_matrix, annot=True, fmt='g', cmap='Blues', 
            xticklabels=grid_svm_gray.classes_,
            yticklabels=grid_svm_gray.classes_)

print(classification_report(y_test, y_test_pred))
````
